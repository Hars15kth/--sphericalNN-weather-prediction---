This notebook visualizes ERA5 500 hPa data and model outputs from baseline CNN and SO(3)-equivariant spherical neural network.

```python
import xarray as xr
import matplotlib.pyplot as plt
from pathlib import Path

ds = xr.open_dataset(Path(r"D:\era5_cache\era5_500hpa_2024_09.nc"), engine="netcdf4")
ds = ds.sel(pressure_level=500.0) if "pressure_level" in ds else ds

temp = ds["t"].isel(time=0)
plt.figure(figsize=(10, 5))
temp.plot(cmap="coolwarm")
plt.title("ERA5 500 hPa Temperature Snapshot")
plt.show()
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt

baseline = np.load(r"D:\era5_cache\runtime\baseline_tgt_64x128.npy")[0, 0]
spherical = np.load(r"D:\era5_cache\runtime\spherical_tgt_CN.npy")[0, 0]

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(baseline, cmap="viridis")
plt.title("Baseline CNN Output")

plt.subplot(1, 2, 2)
plt.plot(spherical[:100])
plt.title("Spherical Model Output (slice)")
plt.tight_layout()
plt.show()
```
